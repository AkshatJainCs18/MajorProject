```python
test=pd.read_csv("C:/Users/ACER/Desktop/Major Project/MajorProject/testing.csv")
```

```python
progdx={}
c=1
for i in test['prognosis']:
    if progdx.get(i)==None:
        progdx[str(i)]=c
        c+=1
        
```

```python
lx=[progdx[i] for i in test['prognosis']]
```

```python
test['progval']=lx
```

```python
yt=test['progval'].values
```

```python
test=test.drop('prognosis',axis=1)
test2=test.drop('progval',axis=1)
```

```python
xt=test2.values
```

```python
from sklearn import tree
dt_clf = tree.DecisionTreeClassifier(max_depth=30)
dt_clf.fit(x,y)
dt_clf_score=dt_clf.score(xt,yt)
target_pred = dt_clf.predict([xt[8]])
```

```python
from sklearn import ensemble

rf_clf = ensemble.RandomForestClassifier(n_estimators=50)
rf_clf.fit(x,y)
rf_clf_score=rf_clf.score(xt,yt)
```

```python
gb_clf = ensemble.GradientBoostingClassifier(n_estimators=35)
gb_clf.fit(x,y)
gb_clf_score=gb_clf.score(xt,yt)
```

```python
from sklearn.naive_bayes import GaussianNB
nb_clf=GaussianNB()
nb_clf.fit(x,y)
nb_clf_score=nb_clf.score(xt,yt)
```

```python
from sklearn.svm import SVC
sv_clf= SVC(probability=True, kernel='linear')
sv_clf.fit(x,y)
sv_clf_score=sv_clf.score(xt,yt)
```

```python
from sklearn.linear_model import LogisticRegression
lr_clf=LogisticRegression()
lr_clf.fit(x,y)
lr_clf_score=lr_clf.score(xt,yt)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn_clf=KNeighborsClassifier(n_neighbors=3)
kn_clf.fit(x,y)
kn_clf_score=kn_clf.score(xt,yt)
```

```python
dt_clf_score,gb_clf_score,nb_clf_score,rf_clf_score,sv_clf_score,lr_clf_score,kn_clf_score
```

```python
classifiers = ["Decision Tree", "Random Forest","KNN","Gradient Boost","Naive Bayes","SVM","Logistic Regression"]
score_values = [dt_clf_score, rf_clf_score,kn_clf_score,gb_clf_score,nb_clf_score,sv_clf_score,lr_clf_score]
fig = plt.figure(figsize = (10,5))
 

plt.bar(classifiers, score_values, color ='maroon',
       width = 0.2)
 
plt.xlabel("Classifier names")
plt.ylabel("R2score")
```
