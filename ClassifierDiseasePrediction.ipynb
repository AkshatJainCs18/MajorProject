```python
xt=test2.values
```

```python
from sklearn import tree
dt_clf = tree.DecisionTreeClassifier(max_depth=100)
dt_clf.fit(x,y)
dt_clf_score=dt_clf.score(xt,yt)
target_pred = dt_clf.predict(xt)
```

```python
target_pred
```

```python
from sklearn import ensemble
rf_clf = ensemble.RandomForestClassifier(n_estimators=34)
rf_clf.fit(x,y)
rf_clf_score=rf_clf.score(xt,yt)
```

```python
classifiers = ["Decision Tree", "Random Forest"]
score_values = [dt_clf_score, rf_clf_score]
fig = plt.figure(figsize = (8,6))
 
# creating the bar plot
plt.bar(classifiers, score_values, color ='maroon',
        width = 0.8)
 
plt.xlabel("Classifier names")
plt.ylabel("R2score")
```
